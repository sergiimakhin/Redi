```python
# Graphic representations of sin/cos pairs to help with intuitive understanding how it works

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with all months (for month plots)
months = np.arange(1, 13)  # 1 to 12
df_months = pd.DataFrame({'month': months})

# Apply cyclic encoding for months
df_months['month_sin'] = np.sin(2 * np.pi * df_months['month'] / 12)
df_months['month_cos'] = np.cos(2 * np.pi * df_months['month'] / 12)

# Plot 1: Raw month vs month_sin and month_cos
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(df_months['month'], df_months['month_sin'], label='month_sin', marker='o')
plt.plot(df_months['month'], df_months['month_cos'], label='month_cos', marker='o')
plt.xlabel('Month')
plt.ylabel('Encoded Value')
plt.title('Cyclic Encoding of Month')
plt.legend()
plt.grid(True)

# Plot 2: month_sin vs month_cos (unit circle)
plt.subplot(1, 2, 2)
plt.scatter(df_months['month_cos'], df_months['month_sin'], c=df_months['month'], cmap='viridis')
for i, month in enumerate(df_months['month']):
    plt.text(df_months['month_cos'][i], df_months['month_sin'][i], str(month))
plt.xlabel('month_cos')
plt.ylabel('month_sin')
plt.title('Month on Unit Circle')
plt.grid(True)
plt.axis('equal')

plt.tight_layout()
plt.savefig('month_encoding.png')

# Create a separate DataFrame for all weekdays (for weekday plots)
weekdays = np.arange(0, 7)  # 0 to 6
df_weekdays = pd.DataFrame({'weekday': weekdays})

# Apply cyclic encoding for weekdays
df_weekdays['weekday_sin'] = np.sin(2 * np.pi * df_weekdays['weekday'] / 7)
df_weekdays['weekday_cos'] = np.cos(2 * np.pi * df_weekdays['weekday'] / 7)

# Plot 3: Raw weekday vs weekday_sin and weekday_cos
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(df_weekdays['weekday'], df_weekdays['weekday_sin'], label='weekday_sin', marker='o')
plt.plot(df_weekdays['weekday'], df_weekdays['weekday_cos'], label='weekday_cos', marker='o')
plt.xlabel('Weekday')
plt.ylabel('Encoded Value')
plt.title('Cyclic Encoding of Weekday')
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.legend()
plt.grid(True)

# Plot 4: weekday_sin vs weekday_cos (unit circle)
plt.subplot(1, 2, 2)
plt.scatter(df_weekdays['weekday_cos'], df_weekdays['weekday_sin'], c=df_weekdays['weekday'], cmap='viridis')
for i, weekday in enumerate(df_weekdays['weekday']):
    plt.text(df_weekdays['weekday_cos'][i], df_weekdays['weekday_sin'][i], str(weekday))
plt.xlabel('weekday_cos')
plt.ylabel('weekday_sin')
plt.title('Weekday on Unit Circle')
plt.grid(True)
plt.axis('equal')

plt.tight_layout()
plt.savefig('weekday_encoding.png')
```
